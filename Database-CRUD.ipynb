```python
cursor.execute('SELECT table_schema AS "Database",     ROUND(SUM(data_length + index_length) / 1024 / 1024, 2) AS "Size (MB)" FROM information_schema.TABLES WHERE table_schema = "CHESSBOT" GROUP BY table_schema;')
records = cursor.fetchall()
for row in records:
    print(row)
```

```python
# 1. Execute the command
cursor.execute("SHOW TABLES")

# 2. Fetch and print the results
print(f"Tables in the database:")
tables = cursor.fetchall()

if not tables:
    print("No tables found.")
else:
    for (table_name,) in tables:
        print(f"- {table_name}")
```

```python
# --- CREATE: Create a simple table ---
cursor.execute("DROP TABLE IF EXISTS test_users")
cursor.execute("""
    CREATE TABLE test_users (
        id INT AUTO_INCREMENT PRIMARY KEY,
        name VARCHAR(100),
        email VARCHAR(100)
    )
""")
print("Table 'test_users' created.")
```

```python
# --- INSERT: Add a record ---
insert_query = "INSERT INTO test_users (name, email) VALUES (%s, %s)"
cursor.execute(insert_query, ("John Doe", "john@example.com"))
connection.commit()  # Required to save changes
print(f"Record inserted. ID: {cursor.lastrowid}")
```

```python
# --- READ: Fetch the record ---
cursor.execute("SELECT * FROM test_users")
records = cursor.fetchall()
print(f"Total rows: {cursor.rowcount}")
for row in records:
    print(f"User: {row[1]} | Email: {row[2]}")
```

```python
# --- UPDATE: Modify the record ---
update_query = "UPDATE test_users SET email = %s WHERE name = %s"
cursor.execute(update_query, ("new_john@example.com", "John Doe"))
connection.commit()
print("Record updated.")
```

```python
# --- DELETE: Remove the record ---
delete_query = "DELETE FROM test_users WHERE name = %s"
cursor.execute(delete_query, ("John Doe",))
connection.commit()
print("Record deleted.")
```
